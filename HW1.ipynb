```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('/home/lycoris/Загрузки/train.csv') 
```

# гипотеза 1: выживаемость зависит от возраста пассажира

```python
ax = data['Age'].hist(figsize=(20, 10), bins=80, edgecolor='white', grid=True, color='pink')

data[data['Survived'] == 0]['Age'].hist(figsize=(20, 10), bins=80, 
                                          edgecolor='white', grid=True, color='grey')
ax.legend(['All passengers', 'Not Survived'], prop={'size':20})
ax.set_xlabel('Age')
ax.set_ylabel('Humans')
```

*количественно погибших больше всего в возрасте 16-40 лет, однако в том же возрасте больше всего пассажиров*

*рассмотрим процент выживших по возрасту*

```python
age_sur = round(100 * data.groupby(['Age'])['Survived'].sum() / data.groupby(['Age'])['PassengerId'].nunique())
age_sur.plot(figsize=(20, 10), linestyle='--', marker='o', markersize=8, grid=True, lw=2, c = 'grey',
             xlabel='Age', ylabel='Precent of survived')
```

итак, несмотря на высокое количество погибших от 16 до 40, процент выживших не так низок. в итоге **однозначной зависимости между возрастом и шансом выжить нет**

# гипотеза 8: некоторые факторы выживаемости взаимосвязаны

## 8.1 взаимосвязь класса пассажира и цены билета

```python
pc_sur = round(data.groupby(['Pclass'])['Fare'].median())
pc_sur.plot(figsize=(10, 5), linestyle='--', marker='o', markersize=8, grid=True, lw=2, c = 'grey',
              xlabel='Pclass', legend='reverse', ylabel='Median fare')
```

То есть медианная цена билета коррелирует с классом и ее также можно брать в расчет выживаемости


## 8.2 взаимосвязь класса пассажира и порта посадки

```python
ax = data[data.Pclass == 3]['Embarked'].hist(figsize=(10, 5), bins=3, edgecolor='white',grid=True, 
                                        color='pink', alpha=0.5)
data[data.Pclass == 2]['Embarked'].hist(figsize=(10, 5), bins=3, edgecolor='white',grid=True, 
                                        color='purple', alpha=0.5)
data[data.Pclass == 1]['Embarked'].hist(figsize=(10, 5), bins=3, edgecolor='white',grid=True, 
                                        color='blue', alpha=0.5)
ax.legend(['3 class', '2 class', '1 class'])
ax.set_xlabel('Embarked')
ax.set_ylabel('Humans')
```

Итак, в основном пассажиры первого класса садились в портах S и C, пассажиры 3 класса - в порту S. 
Этим объясняется странная зависимость выживаемости пассажиров от порта посадки

# гипотеза 9: взаимовыручка и выживание в классе

```python
data['Family'] = data['SibSp'] + data['Parch']
data3 = data[data.Pclass == 3]
data2 = data[data.Pclass == 2]
data1 = data[data.Pclass == 1]
series3 = round(100 * data3.groupby(['Family'])['Survived'].sum() / data3.groupby(['Family'])['Survived'].count())
series3.plot(figsize=(10, 5), linestyle='--', marker='o', markersize=8, lw=2, legend='reverse', c='grey')
series2 = round(100 * data3.groupby(['Family'])['Survived'].sum() / data2.groupby(['Family'])['Survived'].count())
series2.plot(figsize=(10, 5), linestyle='--', marker='o', markersize=8, lw=2, legend='reverse', c='pink')
series1 = round(100 * data3.groupby(['Family'])['Survived'].sum() / data1.groupby(['Family'])['Survived'].count())
series1.plot(figsize=(10, 5), linestyle='--', marker='o', markersize=8, grid=True, lw=2, legend='reverse',
            ylabel='Percent of survived', title='grey is 3class, pink is 2class, purple is 1class', c='purple')
```

Итак, среди пассажиров 3 класса выживали те, кто имел 3 членов семьи, 2 класса - 2, 1 класса - 2 или 3

# гипотеза 9: возраст и выживание в классе

```python
series3 = round(100 * data3.groupby(['Age'])['Survived'].sum() / data3.groupby(['Age'])['Survived'].count())
series3.plot(figsize=(20, 10), linestyle='--', marker='o', markersize=8, lw=2, legend='reverse', c='grey')
series2 = round(100 * data2.groupby(['Age'])['Survived'].sum() / data2.groupby(['Age'])['Survived'].count())
series2.plot(figsize=(20, 10), linestyle='--', marker='o', markersize=8, lw=2, legend='reverse', c='pink')
series1 = round(100 * data1.groupby(['Age'])['Survived'].sum() / data1.groupby(['Age'])['Survived'].count())
series1.plot(figsize=(20, 10), linestyle='--', marker='o', markersize=8, grid=True, lw=2, legend='reverse',
            ylabel='Percent of survived', title='grey is 3class, pink is 2class, purple is 1class', c='purple')
```

Общих закономерностей все еще нет, однако можно заметить, что выжили все дети ~до 15 лет во втором класссе
