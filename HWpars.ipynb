```python
from selenium import webdriver
import time
import pandas as pd
from tqdm import tqdm 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
ptd = '/home/lycoris/python_data_analysis_2021/geckodriver' # path to driver for firefox
```

### Парсинг названий и рейтинга (+ ссылка на рецензии)

```python
base_url = 'https://www.kinopoisk.ru/lists/top250/?page={}&tab=all'
urls = [base_url.format(x) for x in range (1, 6)]
urls # чтобы парсить не только с первой страницы, а со всех 5
```

```python
try:
    
    browser = webdriver.Firefox(executable_path=ptd)  # создаем сессию
    totalt = [] # список списков всех названий
    totalr = [] # список списков всех рейтингов
    totall = [] # список списков всех ссылок на фильмы
    for url in urls:
        browser.get(url)
        titles = browser.find_elements_by_css_selector('p:nth-child(1)')
        rating = browser.find_elements_by_css_selector('div:nth-child(1) > span:nth-child(1) > span:nth-child(1)')
        titles = [i.text for i in titles]
        rating = [i.text for i in rating]
        elems = browser.find_elements_by_class_name("selection-film-item-meta [href]")
        links = [elem.get_attribute('href') for elem in elems]
        links = ['{}/reviews/ord/date/status/all/perpage/200/'.format(link) for link in links]
        totalt.append(titles)
        totalr.append(rating)
        totall.append(links)
    really_totalt = sum(totalt,[]) # итоговый список всех названий
    really_totalr = sum(totalr,[]) # итоговый список всех рейтингов
    really_totall = sum(totall,[]) # итоговый список всех ссылок на рецензии
    print(len(really_totalt)) # проверка
    print(really_totalr) # проверка
    print(len(really_totall)) # проверка
    
except Exception as e:
    
    print('Ошибка')
    print(e)
    
finally:

    time.sleep(1)  
    browser.quit()  
    print('Закрыл')
```

```python
dictio = {'titles':really_totalt, 'rating':really_totalr, 'links':really_totall}
parse_data = pd.DataFrame(dictio) 
```
